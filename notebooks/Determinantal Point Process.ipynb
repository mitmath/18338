## a really nice property of projection processes

```julia
## associate ğ“˜  with its probability
## then associate all j subsets of ğ“˜ with that probability over binomial(p,j)
## now add up the probabilities for all j subsets ... that will be the j subset probabilty
```

```julia
k = 3
d = Dict{Array{Int64,1},Float64}()
 for ğ“˜ âˆˆ combinations(1:N, n)
    prob = det(L[ğ“˜,ğ“˜]) / binomial(N,n)
    for ğ’¬ âˆˆ combinations(ğ“˜)
        if haskey(d,ğ’¬)
            d[ğ’¬] += prob
        else
            d[ğ’¬] = prob
        end
    end
end
```

## The simple thing:
   Take ğ’¬ in the power set of 1:N with probabilty det(L[ğ’¬;ğ’¬])/det(I+L)

## Slightly more complicated:
   If A is the power set of 1:N, the probability that A âŠ† ğ’¬ is det(K[A;A])

what does this mean?
   This means that if we add up all the disjoint situations of the probability of ğ’¬ as above but only when A âŠ† ğ’¬, we will get det(K[A;A])
   
   
   Let's take an example : A ={1}
   What we are going to do is take a random subset of 1:N, and we are wondering what is the probablity that we will see a {1} in this subset.  This probablity is K[1,1].
   This is the sum of P[1] +P[1,2] +P[1,3] + P[1,2,4,5] + lots of other things
   
  what is the probabilty of ğ’¬ = [1] itself?
Answer: it's the probability that [1] shows up , but without any friends.
    it's not K[1;1]. Remember K[1,1] is the probablity that 1 shows up in ğ’¬, usually with friends.  




```julia
println("True Probabilities L form")
println("[]"," ",1/det(I+L))
for ğ’¬ âˆˆ combinations(1:N)
    println(ğ’¬," ",det(L[ğ’¬,ğ’¬])/det(I+L))
end

println()
K = L/(L+I) # L * inv(L+I)
# L = K/(I-K)

println("Cumulatives Probabilities K form")
  println("[] ",1)
for A = combinations(1:N)
# Generate all subsets Y
  
 print(A, " ",sum( det(L[Y,Y]) for Y âˆˆ combinations(1:N) if A âŠ† Y) / det(I+L))
 println(" ",det(K[A,A]) )
end
                
```
