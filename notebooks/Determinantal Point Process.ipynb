## a really nice property of projection processes

```julia
## associate ð“˜  with its probability
## then associate all j subsets of ð“˜ with that probability over binomial(p,j)
## now add up the probabilities for all j subsets ... that will be the j subset probabilty
```

```julia
k = 3
d = Dict{Array{Int64,1},Float64}()
 for ð“˜ âˆˆ combinations(1:N, n)
    prob = det(K[ð“˜,ð“˜]) / binomial(N,n)
    for ð’¬ âˆˆ combinations(ð“˜,k)
        if haskey(d,ð’¬)
            d[ð’¬] += prob
        else
            d[ð’¬] = prob
        end
    end
end
```
