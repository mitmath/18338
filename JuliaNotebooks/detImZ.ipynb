```julia
using SymPy, LinearAlgebra, Combinatorics
```

```julia
# Mz + I
n = 3
z = symbols("z")
A = z.*[ symbols("A_$i$j",real=true) for i=1:n, j=1:n]
M =A + Matrix(I,n,n)
M
```

```julia
d = expand(det(M))
```

```julia
[d.coeff(z^k) for k=n:-1:1]  # find coefficients for z^n down to z^1
```

```julia
for k=n:-1:0

    display(  sum(det(A[𝓘,𝓘])  for 𝓘 ∈ combinations(1:n, k)))
   
end
```
